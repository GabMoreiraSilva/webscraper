```python
import requests 
from bs4 import BeautifulSoup 
    
# Fetch the HTML content 
url = 'https://edge.uol/pt/post-sitemap.xml' 
response = requests.get(url)
html_content = response.content 
    
# Parse the HTML 
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
sitesFiltrados = []
sites = soup.findAll("loc")
for site in sites:
    if "artigo" in site.text:
        sitesFiltrados.append(site.text)
```

```python
sitesFiltrados[:5]
```

```python
import re
dados = []
for site in sitesFiltrados[:5]:
    response = requests.get(site)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    text = soup.get_text(separator='\n', strip=True)
    regex = r"\/([^\/]+)\/$"
    match = re.search(regex, site)
    dados.append({"nome":match.group(1),"data": text})
```

```python
import pandas as pd

df = pd.DataFrame(dados)
```

```python
df.to_csv("./Dados.csv")
```
